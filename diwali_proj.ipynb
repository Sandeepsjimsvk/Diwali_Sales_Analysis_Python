EXPLORATORY DATA ANALYSIS -- (EDA)

GENDER--

```python
# plotting a bar chart for Gender and it's count
ax = sns.countplot(x='Gender', data=df, hue = 'Gender', palette=['#5DADE2', '#58D68D'], legend = False)
for bars in ax.containers: 
    ax.bar_label(bars) 
sns.set(rc={'figure.figsize':(10,5)})
plt.show()
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
palette = {'M': '#5DADE2', 'F': '#58D68D'}
sns.barplot(x = 'Gender', y = 'Amount', hue='Gender', data = sales_gen, palette = palette, legend = False)
plt.show()
```

```python
ax = sns.countplot(data = df, x ='Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()    
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)
palette = { '0-17': '#5DADE2', 
           '18-25': '#58D68D', 
           '26-35': '#F39C12', 
           '36-45': '#E74C3C', 
           '46-50': '#8E44AD', 
           '51-55': '#1ABC9C', 
           '55+': '#2ECC71' 
}
sns.barplot(x ='Age Group', y= 'Amount', hue ='Age Group', data = sales_age, palette= palette, legend = False)
plt.show()
```

```python
# total number of orders from top 10 states
sales_state= df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by= 'Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State', y= 'Orders', hue = 'State',data = sales_state, legend = False)
plt.show()
```

```python
# total amount/sales from top 10 states
sales_state= df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State', y= 'Amount', hue = 'State',data = sales_state, legend = False)
plt.show()
```

```python
palette = ['#5DADE2', '#58D68D']
ax = sns.countplot(data= df, x = 'Marital_Status', hue = 'Marital_Status', palette = palette, legend = False)

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()    
```

```python
sales_status = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)
sns.set (rc= {'figure.figsize':(6,5)})
sns.barplot(data = sales_status, x = 'Marital_Status', y= 'Amount', hue= 'Gender')
plt.show()
```

```python


sns.set(rc={'figure.figsize': (20, 5)})

palette = sns.color_palette("husl", df['Occupation'].nunique())
ax = sns.countplot(data=df, x='Occupation', hue = 'Occupation', palette = palette, legend = False)


for container in ax.containers:
    ax.bar_label(container)


plt.show()
```
